```python
import pandas as pd
```

```python
data=pd.read_csv('prep.csv')
```

```python
data
```

```python
df=pd.get_dummies(data,drop_first=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
X_indep=df.drop('classification_yes',axis=1)
```

```python
Y_dep=df['classification_yes']
```

```python
Y_dep
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_indep,Y_dep,test_size=0.25,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor=DecisionTreeRegressor(random_state=0)
```

```python
regressor
```

```python
regressor.fit(X_train,Y_train)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
sfs_forward = SFS(regressor, 
                  k_features=5,    # number of features to select
                  forward=True,    # forward selection
                  floating=False,  # no floating step
                  scoring='r2',    # metric to optimize
                  cv=5)  
```

```python
sfs_forward
```

```python
sfs_forward.fit(X_indep, Y_dep)
```

```python
print("Selected features (forward):", sfs_forward.k_feature_names_)
```
