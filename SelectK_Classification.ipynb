```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
def select_kbest(indped_X,dep_Y,n):
    kbest=SelectKBest(score_func=chi2,k=n)
    selectKfeature=kbest.fit_transform(indped_X,dep_Y)
    return selectKfeature
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
def split_stdScalar(indped_X,dep_Y):
    X_train,X_test,Y_train,Y_test=train_test_split(indped_X,dep_Y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train,Y_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,Y_train,Y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
def cm_prediction(classifier,X_test,Y_test):
    Y_pred=classifier.predict(X_test)
    cm=confusion_matrix(Y_test,Y_pred)
    acc=accuracy_score(Y_test,Y_pred)
    classReport=classification_report(Y_test,Y_pred)
    return cm,acc,classReport
```

```python
from sklearn.linear_model import LogisticRegression
def logistics(X_train,Y_train,X_test,Y_test):
    classifier=LogisticRegression(random_state=0)
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
from sklearn.svm import SVC
def svm_linear(X_train,Y_train,X_test,Y_test):
    classifier=SVC(kernel='linear',random_state=0)
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
def svm_nonlinear(X_train,Y_train,X_test,Y_test):
    classifier=SVC(kernel='rbf',random_state=0)
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
from sklearn.naive_bayes import GaussianNB
def navibias(X_train,Y_train,X_test,Y_test):
    classifier=GaussianNB()
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
from sklearn.neighbors import KNeighborsClassifier
def KNNclassifier(X_train,Y_train,X_test,Y_test):
    classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
from sklearn.tree import DecisionTreeClassifier
def decisiontree(X_train,Y_train,X_test,Y_test):
    classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
from sklearn.ensemble import RandomForestClassifier
def randomforest(X_train,Y_train,X_test,Y_test):
    classifier=RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    classifier.fit(X_train,Y_train)
    cm,acc,classReport=cm_prediction(classifier,X_test,Y_test)
    return cm,acc,classReport
```

```python
def selectk_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf): 
    
    dataframe=pd.DataFrame(index=['ChiSquare'],columns=['Logistic','SVMl','SVMnl','KNN','Navie','Decision','Random'])
    for number,idex in enumerate(dataframe.index):      
        dataframe['Logistic'][idex]=acclog[number]       
        dataframe['SVMl'][idex]=accsvml[number]
        dataframe['SVMnl'][idex]=accsvmnl[number]
        dataframe['KNN'][idex]=accknn[number]
        dataframe['Navie'][idex]=accnav[number]
        dataframe['Decision'][idex]=accdes[number]
        dataframe['Random'][idex]=accrf[number]
    return dataframe
```

```python
kbest=select_kbest(indped_X,dep_Y,5) 
```

```python
kbest
```

```python
acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]
```

```python
X_train, X_test, Y_train, Y_test=split_stdScalar(kbest,dep_Y) 
```

```python
cm,acc,classReport=logistics(X_train,Y_train,X_test,Y_test)
acclog.append(acc)

cm,acc,classReport=svm_linear(X_train,Y_train,X_test,Y_test)  
accsvml.append(acc)
    
cm,acc,classReport=svm_nonlinear(X_train,Y_train,X_test,Y_test)  
accsvmnl.append(acc)
    
cm,acc,classReport=KNNclassifier(X_train,Y_train,X_test,Y_test)  
accknn.append(acc)
    
cm,acc,classReport=navibias(X_train,Y_train,X_test,Y_test)  
accnav.append(acc)
    
cm,acc,classReport=decisiontree(X_train,Y_train,X_test,Y_test)  
accdes.append(acc)
    
cm,acc,classReport=randomforest(X_train,Y_train,X_test,Y_test)  
accrf.append(acc)
    
result=selectk_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf)

result
```
