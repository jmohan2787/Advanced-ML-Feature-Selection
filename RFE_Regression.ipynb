```python
import pandas as pd
```

```python
data=pd.read_csv('prep.csv')
```

```python
data
```

```python
df=pd.get_dummies(data,drop_first=True)
```

```python
indep_X=df.drop('classification_yes',axis=1)
```

```python
indep_X
```

```python
dep_Y=df['classification_yes']
```

```python
dep_Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
def split_stdScallar(indep_X,dep_Y):
    X_train,X_test,Y_train,Y_test=train_test_split(indep_X,dep_Y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train,Y_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,Y_train,Y_test
```

```python
from sklearn.metrics import r2_score
```

```python
def r2_score_measure(regressor,X_test,Y_test):
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
    
```

```python
from sklearn.linear_model import LinearRegression
def Linear_regression(X_train,Y_train,X_test,Y_test):
    regressor=LinearRegression()
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.svm import SVR
def svm_linear(X_train,Y_train,X_test,Y_test):
    regressor=SVR(kernel='linear')
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.svm import SVR
def svm_nonlinear(X_train,Y_train,X_test,Y_test):
    regressor=SVR(kernel='rbf')
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.tree import DecisionTreeRegressor
def decission(X_train,Y_train,X_test,Y_test):
    regressor=DecisionTreeRegressor(random_state=0)
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.ensemble import RandomForestRegressor
def random(X_train,Y_train,X_test,Y_test):
    regressor=RandomForestRegressor(n_estimators = 10, random_state = 0)
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.feature_selection import RFE
def ref_feature(indep_X,dep_Y,n):
    rfe_results = []
    models = [
        LinearRegression(),
        SVR(kernel='linear'),
        DecisionTreeRegressor(random_state=0),
        RandomForestRegressor(n_estimators=10, random_state=0)
    ]
    for model in models:
        rfe=RFE(estimator=model,n_features_to_select=n)
        selected_features=rfe.fit_transform(indep_X,dep_Y)
        rfe_results.append(selected_features)
    return rfe_results
```

```python
def rfe_regression(acclog,accsvml,accdes,accrf): 
    
    rfedataframe=pd.DataFrame(index=['Linear','SVC','Random','DecisionTree'],columns=['Linear','SVMl',
                                                                                        'Decision','Random'])

    for number,idex in enumerate(rfedataframe.index):
        
        rfedataframe['Linear'][idex]=acclog[number]       
        rfedataframe['SVMl'][idex]=accsvml[number]
        rfedataframe['Decision'][idex]=accdes[number]
        rfedataframe['Random'][idex]=accrf[number]
    return rfedataframe
```

```python
rfelist=ref_feature(indep_X,dep_Y,3)
```

```python
rfelist
```

```python
acclin=[]
accsvml=[]
accsvmnl=[]
accdes=[]
accrf=[]
```

```python
for i in rfelist:   
    X_train, X_test, Y_train, Y_test=split_stdScallar(i,dep_Y)  
    r2_lin=Linear_regression(X_train,Y_train,X_test,Y_test)
    acclin.append(r2_lin)
    
    r2_sl=svm_linear(X_train,Y_train,X_test,Y_test)    
    accsvml.append(r2_sl)
    
    r2_NL=svm_nonlinear(X_train,Y_train,X_test,Y_test)
    accsvmnl.append(r2_NL)
    
    r2_d=decission(X_train,Y_train,X_test,Y_test)
    accdes.append(r2_d)
    
    r2_r=random(X_train,Y_train,X_test,Y_test)
    accrf.append(r2_r)
    
    
result=rfe_regression(acclin,accsvml,accdes,accrf)
```
