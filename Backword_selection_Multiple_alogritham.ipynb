```python
import pandas as pd
```

```python
data=pd.read_csv('prep.csv')
```

```python
data
```

```python
df=pd.get_dummies(data,drop_first=True)
```

```python
X_indep=df.drop('classification_yes', axis=1)
```

```python
X_indep
```

```python
Y_dep=df['classification_yes']
```

```python
Y_dep
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
def split_stad_scallar(X_indep,Y_dep):
    X_train,X_test,Y_train,Y_test=train_test_split(X_indep,Y_dep,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train,Y_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,Y_train,Y_test
```

```python
from sklearn.linear_model import LinearRegression
def Linear_regression(X_train,Y_train,X_test,Y_test):
    regressor=LinearRegression()
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.svm import SVR
def svm_linear(X_train,Y_train,X_test,Y_test):
    regressor=SVR(kernel='linear')
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.svm import SVR
def svm_nonlinear(X_train,Y_train,X_test,Y_test):
    regressor=SVR(kernel='rbf')
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.tree import DecisionTreeRegressor
def decission(X_train,Y_train,X_test,Y_test):
    regressor=DecisionTreeRegressor(random_state=0)
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from sklearn.ensemble import RandomForestRegressor
def random(X_train,Y_train,X_test,Y_test):
    regressor=RandomForestRegressor(n_estimators = 10, random_state = 0)
    regressor.fit(X_train,Y_train)
    Y_pred=regressor.predict(X_test)
    r2=r2_score(Y_test,Y_pred)
    return r2
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
def sfs_feature(indep_X,dep_Y,n):
    sfs_results = []
    models = [
        [LinearRegression(),'LinearRegression'],
        # [SVR(kernel='linear'),'SVR_linear'],
        [DecisionTreeRegressor(random_state=0),'DecisionTree'],
        [RandomForestRegressor(n_estimators=10, random_state=0),'Random_Forest']
    ]
    for model, name in models:
        sfs_forward = SFS(estimator=model, 
                          k_features=n,
                          forward=False,
                          floating=False,
                          scoring='r2',
                          cv=5,
                          n_jobs=-1)
        
        selected_features=sfs_forward.fit_transform(indep_X, dep_Y)
        selected_columns = indep_X.columns[list(sfs_forward.k_feature_idx_)]
        r2_cv_score = sfs_forward.k_score_

        print(f"Model: {name}")
        print(f"Selected Columns: {list(selected_columns)}")
        print(f"Cross-Validated R² Score: {r2_cv_score:.4f}")

        sfs_results.append(selected_features)

    return sfs_results
```

```python
def sfs_regression(acclog,accdes,accrf): 
    
    rfedataframe=pd.DataFrame(index=['Linear','Random','DecisionTree'],columns=['Linear',
                                                                                        'Decision','Random'])

    for number,idex in enumerate(rfedataframe.index):
        
        rfedataframe.loc[idex, 'Linear'] = acclog[number]
     
        rfedataframe.loc[idex, 'Decision'] = accdes[number]
        rfedataframe.loc[idex, 'Random'] = accrf[number]
    return rfedataframe
```

```python
sfslist=sfs_feature(X_indep,Y_dep,3)
```

```python
sfslist
```

```python
acclin=[]
accsvml=[]
accsvmnl=[]
accdes=[]
accrf=[]
```

```python
for entry in sfslist:
    

    X_train, X_test, Y_train, Y_test = split_stad_scallar(entry, Y_dep)

    r2_lin = Linear_regression(X_train, Y_train, X_test, Y_test)
    acclin.append(r2_lin)

    r2_sl = svm_linear(X_train, Y_train, X_test, Y_test)
    accsvml.append(r2_sl)

    r2_NL = svm_nonlinear(X_train, Y_train, X_test, Y_test)
    accsvmnl.append(r2_NL)

    r2_d = decission(X_train, Y_train, X_test, Y_test)
    accdes.append(r2_d)

    r2_r = random(X_train, Y_train, X_test, Y_test)
    accrf.append(r2_r)

# Final aggregation (example function)
result = sfs_regression(acclin, accdes, accrf)
```
